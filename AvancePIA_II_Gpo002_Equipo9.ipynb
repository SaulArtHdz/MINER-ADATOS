##### GRÁFICAS.

```python
# Queremos ver la proporción total de nuestra base de datos respecto a la información. 
(df_NoNBA_UltTemp['League'].value_counts() / len(df_NoNBA_UltTemp)).plot.bar()
```

De la gráfica anterior, podemos hacer la suposición de que la liga Chinese - CBA cuenta con más jugadores que las demás ligas en la última temporada, por lo que podríamos decir que la competitividad prodría presentarse más para esta liga que para las restantes de la base de datos.

```python
#Distribución de los juegos jugados de la última temporada.
df_NoNBA_UltTemp[df_NoNBA_UltTemp['GP'] < 80]['GP'].plot.hist()  
```

Dentro de un histograma podemos ver que la moda se encuentra entre los 10 y 20 partidos jugados, a más partidos jugados, menos frecuencia de los jugadores que han participado en más de 30 juegos, gráficamente podemos ver que puede seguir una distribución normal. 

```python
#Distribución de los goles de campos hecho de la última temporada.
df_NoNBA_UltTemp['FGM'].plot.hist() 
```

Para la distribución de los goles de campo, vemos que la moda se encuentra en las menores cantidad, por lo que podríamos decir que solo pocos son los que consiguen destacar en cuanto a ser un anotador se refiere, puede influír mucho la posición en la que se sitúe el jugador, para ello, necesitamos comparar para las demás estadísticas. 

```python
#Relación entre los juegos jugados y los goles de campo hechos. 
df_NoNBA_UltTemp[df_NoNBA_UltTemp['GP'] < 80].plot.scatter(x='GP', y='FGM')  
```

Podemos ver que entre más juegos jugados tiene el jugador, más puntos son anotados, podemos notar una concentración que con la siguiente gráfica podemos distinguir más. 

```python
#La concentración se encuentra en menos juegos y menos puntos.
df_NoNBA_UltTemp[df_NoNBA_UltTemp['GP'] < 80].plot.hexbin(x='GP', y='FGM', gridsize=10) 
```
